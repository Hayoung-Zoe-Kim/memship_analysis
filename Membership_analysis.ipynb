# Membership data analysis 
## - Exploratory analysis & Visualization

##### - about the data
The data set used in below analysis is a new membership data set of one fashion e-commerce platform.

*data : 'Membership_2022.csv' 
- It only contains membership who newly enrolled in 2022 and who has at least once visited the website in 2022
- It had been collected from the beginnig of 2022 until the end of July 2022.
- Moderating the size of data set, it only contains the data of new members who at least for once purchased the proudcts.

##### - about the columns
Since most of columns are explicitly named, I minized the column description

- membership_grade: five grades exist and the grade will be upgraded to next level according to the accumulated spending amount follwing the order as written - Bronze, Silver, Gold, Platinum, Diamond
- marketing_message_Y: whether the members agree to get marketing messages or not
- purchase_amount: the amount of products that the members had bought in 2022 (until July)
- payment_amount: the amount of accumulated spending that the members had spent in 2022 (until July)

### Repurchase_Y/N (N means that members purchased only for once)

```python
#add a new column 'repurchase' that indicates whether the members repurchasd or not
def repurchase(x):
    if x > 1:
        return 'Y'
    else:
        return 'N'
```

```python
data['repurchase'] = data['purchase_amount'].copy()
```

```python
data['repurchase'] = data['repurchase'].apply(func = repurchase)
```

```python
plt.figure(figsize = (10,5))
sns.countplot(x='repurchase', data=data, palette = 'coolwarm', hue = 'gender')
```

```python
data[data['purchase_amount'] > 1]['purchase_amount'].count() / data['purchase_amount'].count()
```

```python
data['purchase_amount'].sum() / data['purchase_amount'].count()
```

# 7. The Correlation between Activities (sign-up, log-in, purchase, repurchase)

```python
#convert date data to the datetime type
from datetime import datetime, timedelta
```

```python
data['recent_login_date'] = data['recent_login_date'].apply(lambda x: pd.to_datetime(str(x),format='%Y%m%d'))
data['recent_purchase_date'] = data['recent_purchase_date'].apply(lambda x: pd.to_datetime(str(x),format='%Y%m%d'))
data['Membership_startdate'] = data['Membership_startdate'].apply(lambda x: pd.to_datetime(str(x),format='%Y%m%d'))
```

### Recent log-in date & purchase date

```python
plt.figure(figsize = (5,10))
plt.gcf().autofmt_xdate()

sns.jointplot(x = 'recent_login_date', y='recent_purchase_date', data = data, hue ='marketing_message_Y', size = 8, palette = 'coolwarm', hue_order = ['Y','N'])
```

```python
plt.figure(figsize = (5,10))
plt.gcf().autofmt_xdate()

sns.jointplot(x = 'recent_login_date', y='recent_purchase_date', data = data, hue ='repurchase', size = 8,hue_order = ['Y','N'])
```

### Membership start date & Recent "Purchase" date, combining with the repurchase status

```python
plt.figure(figsize = (5,10))
plt.gcf().autofmt_xdate()

sns.jointplot(x = 'Membership_startdate', y='recent_purchase_date', hue = 'repurchase', data = data, size = 8)
```

```python
(data['Membership_startdate'] == data['recent_purchase_date']).sum() / data['Membership_startdate'].count()
```

### The length of Membership & Recent activities (log-in, purchase)

```python
today = datetime.now()
```

```python
time_diff = today - data['recent_login_date'].loc[0]
time_diff = time_diff / timedelta(days=1)
```

```python
data['days_after_last_login'] = data['recent_login_date'].apply(lambda x: (today - x) / timedelta(days=1))
data['days_after_last_purchase']= data['recent_purchase_date'].apply(lambda x: (today - x) / timedelta(days=1))
```

```python
data['membership_length'] = data['Membership_startdate'].apply(lambda x : (today - x) / timedelta(days=1))
```

```python
sns.jointplot(x = 'days_after_last_login', y='days_after_last_purchase', data = data, hue = 'membership_length', size = 8, palette = 'coolwarm')
```

### The length of Membership & accumulated spending amount

```python
sns.displot(data=data, x="membership_length", y="payment_amount", hue = 'gender', height = 8)
```

### Finding : 

### - As can be expected, the recent log-in date and recent purchase date shows a positive correlation (although it is not expresesd in a strict inclination)

### - Simultaneously, simple trials of log-in mostly failed to lead to the purchase conversion, referring to the majority of recent purchase dates failed to keep pace with recent log-in dates.

### - In terms of 'marketing message (advertising) acceptance', the members who allowed to accept advertising messages showed stronger activation in both log-in and purchase histories.

### - It turned out that about 61% of members took a first step of purchase at their membership sign-up date, which can be equivalently interpreted that quite a lot of people sign up for membership in the face of the first purchase.
